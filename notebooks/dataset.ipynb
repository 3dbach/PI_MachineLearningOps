```python
# Cargar el dataframe final
final_df = pd.read_csv('dataset_dos_limpio.csv')

# Definir la función userdata
def userdata(User_id: str):
    # Filtrar las revisiones del usuario
    user_reviews = final_df[final_df['user_id'] == User_id]

    # Calcular el dinero gastado por el usuario
    # Convertir la columna de precio a float y manejar casos donde el precio es NaN o 'Free To Play'
    user_reviews['price'] = user_reviews['price'].replace(['Free To Play', 'Free to Play'], 0)
    user_reviews['price'] = pd.to_numeric(user_reviews['price'], errors='coerce').fillna(0)
    total_spent = user_reviews['price'].sum()

    # Calcular el porcentaje de recomendación
    total_reviews = len(user_reviews)
    recommended_reviews = user_reviews['recommend'].sum()
    recommendation_percentage = (recommended_reviews / total_reviews) * 100 if total_reviews else 0

    # Calcular la cantidad de items del usuario
    total_items = user_reviews['items_count'].iloc[0] if not user_reviews.empty else 0

    return {
        "Usuario": User_id,
        "Dinero gastado": f"{float(total_spent)} USD",   # Convertir explícitamente a float
        "% de recomendación": f"{float(recommendation_percentage):.2f}%",  # Convertir explícitamente a float
        "cantidad de items": int(total_items)  # Convertir explícitamente a int
    }

# Probar la función con un User_id de ejemplo
test_user_id = "Skillbot123"
userdata(test_user_id)
```

```python
def userdata_fixed(User_id: str):
    # Filtrar las revisiones del usuario
    user_reviews = final_df[final_df['user_id'] == User_id].copy()  # Usar .copy() para evitar la advertencia

    # Calcular el dinero gastado por el usuario
    # Convertir la columna de precio a float y manejar casos donde el precio es NaN o 'Free To Play'
    user_reviews.loc[:, 'price'] = user_reviews['price'].replace(['Free To Play', 'Free to Play'], 0)
    user_reviews.loc[:, 'price'] = pd.to_numeric(user_reviews['price'], errors='coerce').fillna(0)
    total_spent = user_reviews['price'].sum()

    # Calcular el porcentaje de recomendación
    total_reviews = len(user_reviews)
    recommended_reviews = user_reviews['recommend'].sum()
    recommendation_percentage = (recommended_reviews / total_reviews) * 100 if total_reviews else 0

    # Calcular la cantidad de items del usuario
    total_items = user_reviews['items_count'].iloc[0] if not user_reviews.empty else 0

    return {
        "Usuario": User_id,
        "Dinero gastado": f"{float(total_spent)} USD",   # Convertir explícitamente a float
        "% de recomendación": f"{float(recommendation_percentage):.2f}%",  # Convertir explícitamente a float
        "cantidad de items": int(total_items)  # Convertir explícitamente a int
    }

# Probar la función corregida con un User_id de ejemplo
userdata_fixed(test_user_id)
```

```python
def UserForGenre_test(genero: str):
    # Filtrar los juegos que pertenecen al género especificado
    games_in_genre = steam_games_cleaned[steam_games_cleaned['genres'].str.contains(genero, na=False, case=False)]
    
    # Vincular los juegos filtrados con reduced_items_cleaned para obtener las horas jugadas de cada juego por cada usuario
    user_playtime = games_in_genre.merge(items_cleaned, left_on='id', right_on='item_id', how='inner')
    
    # Agrupar por user_id y sumar las horas jugadas para encontrar el usuario con más horas jugadas
    user_total_playtime = user_playtime.groupby('user_id')['playtime_forever'].sum().reset_index()
    top_user = user_total_playtime.sort_values(by='playtime_forever', ascending=False).iloc[0]['user_id']
    
    # Agrupar por release_date y sumar las horas jugadas para cada año
    hours_by_year = user_playtime.groupby(user_playtime['release_date'].dt.year)['playtime_forever'].sum().reset_index()
    hours_by_year = hours_by_year.rename(columns={"release_date": "Año", "playtime_forever": "Horas"})
    
    # Convertir el DataFrame a una lista de diccionarios
    hours_list = hours_by_year.to_dict('records')
    
    return {
        "Usuario con más horas jugadas para Género {}".format(genero): top_user,
        "Horas jugadas": hours_list
    }

# Testear la función para el género "Action"
test_result = UserForGenre_test("Action")
test_result
```

```python
from textblob import TextBlob

def sentiment_analysis(text):
    # Si la reseña está ausente, retorna Neutral (1)
    if pd.isna(text):
        return 1
    
    # Calcula la polaridad de la reseña
    polarity = TextBlob(text).sentiment.polarity
    
    # Clasifica la reseña según la polaridad
    if polarity < -0.05:
        return 0  # Malo
    elif polarity > 0.05:
        return 2  # Positivo
    else:
        return 1  # Neutral

# Aplicando la función de análisis de sentimiento al dataset user_reviews_df
user_reviews_df4['sentiment_analysis'] = user_reviews_df4['review'].apply(sentiment_analysis)

# Mostrando las primeras filas con la nueva columna
user_reviews_df4[['review', 'sentiment_analysis']].head()
```

```python
def best_developer_year_updated(año: int, df=merged_data):
    # Filtrar los juegos que fueron lanzados en el año especificado
    games_of_year = df[df['release_date'] == año]
    
    # Filtrar las revisiones que tienen recommend como True
    recommended_reviews = games_of_year[games_of_year['recommend'] == True]
    
    # Agrupar por developer y contar las recomendaciones
    developer_counts = recommended_reviews.groupby('developer').size().reset_index(name='recommendations')
    top_developers = developer_counts.sort_values(by='recommendations', ascending=False).head(3)
    
    # Preparar la lista de resultados
    results = [{"Puesto {}".format(i+1): dev} for i, dev in enumerate(top_developers['developer'])]
    
    return results

# Probar la función para el año 1998
best_developer_year_updated(1998)
```

```python
def developer_reviews_analysis(desarrolladora: str, df=merged_data):
    # Filtrar los registros por el desarrollador especificado
    developer_data = df[df['developer'] == desarrolladora]
    
    # Contar los registros de análisis de sentimiento positivo y negativo
    negative_count = len(developer_data[developer_data['sentiment_analysis'] == 0])
    positive_count = len(developer_data[developer_data['sentiment_analysis'] == 1])
    
    # Preparar el resultado
    result = {desarrolladora: [f"Negative = {negative_count}", f"Positive = {positive_count}"]}
    
    return result

# Probar la función para el desarrollador "Valve"
developer_reviews_analysis("Valve")
```

```python
import pandas as pd

# Leer el archivo CSV
dfdf = pd.read_csv("dataset_cuatro.csv")

# Mostrar las primeras filas del dataframe
dfdf.head()
```

```python
def best_developer_year_corrected(dfdf, año):
    # Filtrar el dataframe por el año dado y donde se recomienda el juego y el análisis de sentimiento es positivo
    filtered_df = dfdf[(dfdf['release_date'] == año) & (dfdf['recommend'] == True) & (dfdf['sentiment_analysis'] == 2)]
    
    # Agrupar por desarrollador y contar el número de recomendaciones
    grouped_df = filtered_df.groupby('developer').size().reset_index(name='recommend_count')
    
    # Ordenar los desarrolladores por el número de recomendaciones en orden descendente y obtener los primeros 3
    top_3 = grouped_df.sort_values(by='recommend_count', ascending=False).head(3)
    
    # Convertir el resultado a la estructura deseada
    result = [{"Puesto {}".format(i+1): row.developer} for i, row in enumerate(top_3.itertuples())]
    
    return result

# Probemos la función corregida con el mismo año de ejemplo
best_developer_year_corrected(dfdf, 1998)
```
