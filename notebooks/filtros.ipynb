```python
from pandas import json_normalize

# Cargar todo el archivo JSON en una estructura de datos de Python
with open("australian_user_reviews.json", "r", encoding="utf-8") as file:
    json_content = [eval(line) for line in file]

# Usar json_normalize para desanidar la estructura
normalized_df = json_normalize(json_content, record_path='reviews', meta=['user_id', 'user_url'])

# Mostrar las primeras filas del DataFrame normalizado para verificar
normalized_df.head()
```

```python
# Guardando el DataFrame como CSV
normalized_df.to_csv('user_reviews.csv', index=False)
```

```python
import pandas as pd

# Cargar el archivo CSV
df_genres_tags = pd.read_csv('genres_tags.csv')

# Mostrar las primeras filas del DataFrame
df_genres_tags.head()
```

```python
import ast

# Convertir las cadenas de texto que representan listas en listas reales
df_genres_tags['genres'] = df_genres_tags['genres'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else [])
df_genres_tags['tags'] = df_genres_tags['tags'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else [])

# Desanidar la columna genres
df_genres_expanded = df_genres_tags.explode('genres')

# Mostrar las primeras filas del DataFrame desanidado
df_genres_expanded.head(10)
```

```python
df_genres_expanded
```

```python
# Desanidar la columna tags
df_tags_expanded = df_genres_expanded.explode('tags')

# Mostrar las primeras filas del DataFrame completamente desanidado
df_tags_expanded.head(20)
```

```python
import pandas as pd
import gzip
import json

def parse(path):
    with gzip.open(path, 'r') as g:
        for l in g:
            yield eval(l)
rows = []

for line in parse('users_items.json.gz'):
    row = pd.json_normalize(line)
    rows.append(row)

items_df = pd.concat(rows, ignore_index = True)
```

```python
items_df
```

```python
items_df.to_csv('items.csv', index=False)
```

```python
items= items_df.explode('items')
items = pd.json_normalize(items['items'])
```
